# 2.Problèmatique

Les fournisseurs de fruits et légumes fournissent des périodes de promotion sur quelques produits pour garantir des prix raisonnables aux clients.
Ces fournisseurs proposent certainement des prix différents pour un même produit dans le cadre de concurrence. Dans ce cas, le client (user) cherche comment profiter de cette période face à cette guerre de prix. 

La problèmatique qui s'oppose dans le cadre de notre projet cherche à guider les clients pour mieux profiter des prédiores de promotions avec des meilleurs prix.

Quel est le meilleur moment pour commander à prix bas?

# 4.Réponse à la problématique

Notre solution est un bon plan pour bénéficier des meilleures promotions, acheter moins cher et doubler vos chiffres d'affaires. C'est un guide qui vous facilite la sélection des produits chez le fournisseur de votre choix.

Par exemple dans notre cas, le magasin peut profiter des promotions de Pomme Granny Smith chez le fournisseur F1 et faire son stock.

D'autre part, on trouve des promotions sur le produit Pomme Golden Delicious chez le fournisseur F2 (1.5€ le kilo)entre le mois Juin et le mois de Juillet ainsi que chez le fournisseur F5 entre Juillet et Août (2€ le kilo). Donc ici, à l'aide de nos visualisations inréractives le user trouve facilement le meilleur prix avec lequel peut faire son stock.
De même, les fournisseurs F3 e F4 font des une promotion sur le produit Orange Valence avec deux prix différents(2.5€ chez F3 et 3€ chez F4 )

Dans un second temps, notre plateforme va intégrer des critères supplimentaires très pertinents pour vous aider à la prise de décision tels que l'origine du produit, la qualité du produit et la catégorie.

Notre solution fournit un gain du temps efficace au lieu de passer  probablement beaucoup de temps à réfléchir aux moyens d'économiser de l'argent lorsque vous commader vos produits et choisir la meilleur période pour bénéficier de ces réductions.
