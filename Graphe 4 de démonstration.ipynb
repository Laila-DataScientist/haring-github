# 3.Jeu de données

```python
import pandas as pd
sales_file = pd.ExcelFile("/home/lbenromdhane/Téléchargements/U-Express-Tombe-Issoire-Ventes-jour-ORIGINAL-01082020-au-08102020-(Greenvillage-10oct2020).xlsx")
```

```python
def get_dataframe(sheet_name):
    
    df = pd.read_excel(sales_file, sheet_name, header = None)
    
    # Construction du dataframe
    all_df = pd.DataFrame()
    all_dates = []
    
    number_of_dates = 12
    for i in range(number_of_dates+1):

        df_tmp = df[[0, 1, 4*i+2, 4*i+3, 4*i+4, 4*i+5]]
        period = df_tmp.iloc[0].unique()[1]
        all_dates.append(period)
        
        df_tmp.columns = df_tmp.iloc[1]
        df_tmp = df_tmp.iloc[2:]
        df_tmp['Date'] = period

        all_df = pd.concat([all_df, df_tmp])

    # Suppression des lignes TOTAL GENERAL, FRUITS and LEGUMES
    all_df = all_df.loc[~all_df['EAN-Libellé'].isin(['TOTAL GENERAL', 'FRUITS', 'LEGUMES'])].copy()
    
    # Suppression des lignes cumul
    all_df = all_df.loc[all_df['Date'] != 'Cumul'].copy()

    # Séparation des libéllés
    all_df['EAN'], all_df['Libellé'] = all_df['EAN-Libellé'].str.split(' ', 1).str
    
    del all_df['EAN-Libellé']
    
    #print(sheet_name, all_dates)
    return all_df
```

```python
dataset = pd.DataFrame()

for sheet in range(1, 7):    
    dataset = pd.concat([dataset, get_dataframe(str(sheet))], sort = False)
    
dataset['Date'] = (dataset['Date'] + '20').apply(lambda x: x.strip())
dataset['Date'] = pd.to_datetime(dataset['Date'], format = '%d/%m/%Y')
dataset = dataset[['EAN', 'Libellé', 'Date', 'Nature', 'CA TTC', 'Nb. articles', 'Tx Mrg Brute', 'Quantité vendue']]

# Garder les quantités totales vendues en promotion ou hors promotion
dataset = dataset.loc[dataset.Nature.isnull()].copy() 
del dataset['Nature']

# Supprimer les lignes dupliquées et les lignes ayant 0 articles
dataset.drop_duplicates(inplace = True)
dataset = dataset.loc[dataset['Nb. articles'] != 0]

#transformer en minuscule/ Supprimez toutes les ponctuations .replace('[^\w\s]','') / Supprimer les nombres 
dataset['Libellé_miniscule'] = dataset['Libellé'].str.lower().str.replace('\d+', '')

dataset['Quantité vendue'] = (dataset['Quantité vendue']*10).astype(float)
dataset['Nb. articles'] = dataset['Nb. articles'].astype(float)  

dataset=dataset.reset_index(drop=True)
```

```python
df_fruit = pd.DataFrame()
for produit in dataset['Libellé'].unique():
    df_prod =  dataset.loc[dataset['Libellé'] == produit]
    df_fruit = pd.concat([df_fruit,df_prod])[['Date','Libellé','Quantité vendue']].reset_index(drop = True)
df_fruit.rename(columns = {'Date':'ds','Quantité vendue': 'y'},inplace = True)
df_fruit = df_fruit[0:]
df_fruit
```

On a crée une base de donnée associée à chaque produit dont l'objetctif et de prédire le volume optimal de commande de chaque produit.
On trouve séparement l'historique de commande des produits suivants qu'on pris dans note exemple : POMME GRANNY SMITH, NECTARINE JAUNE, NECTARINE BLANCHE et PECHE JAUNE.

ds :  modèlise la colonne des dates
y : modèlise la quantité de commande 

```python
def splitframe(data, name='name'):
    
    n = data[name][0]

    df = pd.DataFrame(columns=data.columns)

    datalist = []

    for i in range(len(data)):
        if data[name][i] == n:
            df = df.append(data.iloc[i])
        else:
            datalist.append(df)
            df = pd.DataFrame(columns=data.columns)
            n = data[name][i]
            df = df.append(data.iloc[i])
        
    return datalist
```

```python
#affichage
for j in range(4):
    for cluster in splitframe(df_fruit, name = 'Libellé')[j]['Libellé'].unique():
        
        print('produit: ', cluster)
        display(splitframe(df_fruit, name = 'Libellé')[j][['ds', 'Libellé','y']])
```

Prédiction

```python
from fbprophet import Prophet
results = []
prediction = pd.DataFrame()
for j in range(4):
    m = Prophet()
    m.fit(pd.DataFrame(splitframe(df_fruit,name = 'Libellé')[j]))
    future = m.make_future_dataframe(periods=30,include_history = False)
    result = m.predict(future)
    results.append(result)
    result['produit'] = splitframe(df_fruit,name = 'Libellé')[j].iloc[1,1]
    result = result[['ds','produit','yhat', 'yhat_lower', 'yhat_upper' ]]
    prediction = pd.concat([prediction,result]).reset_index(drop = True)

prediction
```
